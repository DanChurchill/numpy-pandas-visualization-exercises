```python
import matplotlib.pyplot as plt
import math
```

```python
# Make a new Jupyter notebook named big_o_notation.ipynb

# Title your chart "Big O Notation"
# Label your x axis "Elements"
# Label your y axis "Operations"
# Label your curves or make a legend for the curves
# Use LaTex notation where possible

y1 = [0 * x + 1 for x in range(0,100)]
y2 = [math.log(x) for x in range(1,100)]
y3 = [x for x in range (0,100)]
y4 = [x * math.log(x) for x in range(1,50)]
y5 = [x**2 for x in range(0,15)]
y6 = [2**x for x in range(0,9)]
y7 = [math.factorial(x) for x in range(0,7)]
y8 = [x^x for x in range(0,100)]

plt.plot(y1, c='green', label='$O(1)$')
plt.plot(y2, c='red', label='$O(log n)$')
plt.plot(y3, c='black', label='$O(n)$')
plt.plot(y4, c='blue', label='$O(n log n)')
plt.plot(y5, c='orange', label='$O(n^2)$')
plt.plot(y6, c='cyan', label='$O(2^n)$')
plt.plot(y7, c='magenta', label='$O(n!)$')
plt.plot(y8, c='yellow', label='$O(n^n)$')

plt.ylim(0, 125)

plt.title('Big O Notation')
plt.legend(loc='upper right')
plt.xlabel('$Elements$')
plt.ylabel('$Operations$')

plt.show()
```
