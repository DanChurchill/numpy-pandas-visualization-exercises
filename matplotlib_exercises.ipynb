```python
import matplotlib.pyplot as plt
import math
```

```python
# Use matplotlib to plot the following equation: y = x**2 - x + 2
# You'll need to write the code that generates the x and y points.
# Add an anotation for the point 0, 0, the origin.

x = range(-100,100)
y = [(n**2 - n + 2) for n in x]
plt.plot(x,y)
plt.title('A plot of: $y = x^2$')
plt.annotate('Origin', xy=(0,0), xytext=(-25,2500), arrowprops={'facecolor':'red'})
plt.show()
```

```python
# y=√x
x = range(0,100)
y = [math.sqrt(n) for n in x]
plt.plot(x,y)
plt.title('A plot of: $y=\sqrt{2}$')
plt.show()
```

```python
# y=x^3
x = range(-100,100)
y = [n**3 for n in x]
plt.plot(x,y)
plt.title('A plot of: $x^3$')
plt.show()
```

```python
# y=2^x
x = range(0,100)
y = [2**n for n in x]
plt.plot(x,y)
plt.title('A plot of: $2^x$')
plt.show()
```

```python
# y=1/(x+1)
x = range(0,100)
y = [1/(n+1) for n in x]
plt_1 = plt.figure(figsize=(10, 10))
plt.plot(x,y)
plt.title('A plot of: $1/(x+1)$')

plt.show()
```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.

x1 = [math.sqrt(n) for n in range(0,100)]
x2 = [x**3 for x in range(-100,100)]
x3 = [2**x for x in range(0,100)]
x4 = [1/(x+1) for x in range(0,100)]
plt.figure(figsize = (10,10))


plt.subplot(221)
plt.plot(x1)
plt.title('$y=\sqrt{2}$')

plt.subplot(222)
plt.plot(x2)
plt.title('$x^3$')

plt.subplot(223)
plt.plot(x3)
plt.title('$2^x$')

plt.subplot(224)
plt.plot(x4)
plt.title('$1/(x+1)$')

plt.show()

```

```python
# Combine the figures you created in the last step into one large figure with 4 subplots.
# Combine the figures you created in the last step into one figure
# where each of the 4 equations has a different color for the points. 
# Be sure to include a legend and an appropriate title for the figure.

plt.figure(figsize=(10, 10))
x1 = [math.sqrt(n) for n in range(0,100)]
x2 = [x**3 for x in range(-100,100)]
x3 = [2**x for x in range(0,100)]
x4 = [1/(x+1) for x in range(0,100)]

plt.plot(x1, c='green', label='$y=\sqrt{2}$')
plt.plot(x2, c='red', label='$x^3$')
plt.plot(x3, c='blue', label='$2^x$')
plt.plot(x4, c='black', label='$1/(x+1)$')

plt.xlim(-10,150)
plt.ylim(0,10)

plt.title('A plot of Four functions')
plt.legend(loc='upper right')
plt.show()

```
